```python
import spacy
import textacy


# Cargar modelo de SpaCy
nlp = spacy.load("en_core_web_sm")

# Función para extraer palabras clave de un texto utilizando Textacy
def extraer_palabras_clave(texto):
    doc = nlp(texto)
    # Utilizamos TextRank para extraer palabras clave
    keywords = textacy.extract.keyterms.textrank(doc)
    return [keyword[0].lower() for keyword in keywords]

# Lista de textos
textos = news.iloc[:,1]

# Extraer palabras clave para cada texto en la lista
palabras_clave_por_texto = [extraer_palabras_clave(texto) for texto in textos]

# Mostrar las palabras clave para cada texto
for i, palabras_clave in enumerate(palabras_clave_por_texto):
    print(f"Palabras clave para el texto {i + 1}: {palabras_clave}")

```

NER con heurísticas

```python
# Lista de ciudades australianas (lexicón)
ciudades_australianas = [
    "sydney", "melbourne", "brisbane", "perth", "adelaide", "canberra",
    "hobart", "darwin", "gold coast", "newcastle", "wollongong", "australia"
]
```

```python
import re

# Definir una función para detectar ciudades australianas en un texto
def detectar_ciudades(texto):
    doc = en(texto)
    ciudades_detectadas = []

    # Verificar si algún token coincide con el lexicón de ciudades
    for token in doc:
        if token.text in ciudades_australianas and token.pos_ in ["PROPN"]:
            ciudades_detectadas.append(token.text)

    # Usar una expresión regular para buscar posibles menciones de ciudades
    regex = re.compile(r'\b(?:' + '|'.join(ciudades_australianas) + r')\b')
    coincidencias = regex.findall(texto)

    # Combinar y deduplicar resultados
    todas_ciudades = list(set(ciudades_detectadas + coincidencias))

    return todas_ciudades
```

```python
textos = news.iloc[0:40000,1]

# Detectar ciudades en cada titular
for i, titular in enumerate(textos):
    ciudades = detectar_ciudades(titular)
    if ciudades:  # Solo imprimir si se detectaron ciudades
        print(f"Título {i + 1}: {titular}")
        print(f"Ciudades detectadas: {ciudades}\n")
```

```python
import spacy

# Cargar modelo de SpaCy
nlp = spacy.load("en_core_web_sm")

# Función para extraer entidades nombradas de un texto utilizando SpaCy
def extraer_entidades(texto):
    doc = nlp(texto)
    entidades = [(ent.text, ent.label_) for ent in doc.ents]
    return entidades

# Lista de textos
textos = news.iloc[:,1]


# Extraer entidades nombradas para cada texto en la lista
entidades_por_texto = [extraer_entidades(texto) for texto in textos]

# Mostrar las entidades nombradas para cada texto
for i, entidades in enumerate(entidades_por_texto):
    print(f"Entidades nombradas para el texto {i + 1}: {entidades}")
```

```python
from datasets import load_dataset

# Cargar el dataset
raw_datasets = load_dataset('nickmuchi/financial-classification')

# Mostrar un ejemplo del dataset para entender su estructura
print(raw_datasets)
print(raw_datasets["train"][0].keys())
```

```python
from transformers import AutoTokenizer, DataCollatorWithPadding


raw_train_dataset = raw_datasets["train"]
print(raw_train_dataset[0])

checkpoint = "bert-base-uncased" # Elegimos transformer encoder
tokenizer = AutoTokenizer.from_pretrained(checkpoint)

# Creamos una función para tokenizar cómodamente
def tokenize_function(example):
    return tokenizer(example["text"], truncation=True)

# Mostramos un ejemplo de texto tokenizado
tokenized_datasets = raw_datasets.map(tokenize_function, batched=True)
print(tokenized_datasets["train"][0])

# Creamos el data_collator
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments("test-trainer")
```

```python
from transformers import AutoModelForSequenceClassification

# Construimos un modelo de clasificación a partir de encoder seleccionado "bert-base-uncased", empleamos la función más abstracta de HuggingFace con valores por defecto
model = AutoModelForSequenceClassification.from_pretrained(checkpoint, num_labels=3)
```

```python
from transformers import Trainer

# Creamos el objeto para entrenar más sencillo proporcionado por HuggingFace
trainer = Trainer(
    model,
    training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["test"],
    data_collator=data_collator,
    tokenizer=tokenizer,
)
```

```python
trainer.train()
```

```python
# Creamos las predicción sobre el modelo entrenado
predictions = trainer.predict(tokenized_datasets["test"])
print(predictions.predictions.shape, predictions.label_ids.shape)
```

```python
import numpy as np

preds = np.argmax(predictions.predictions, axis=-1)
```

```python
import evaluate

metric = evaluate.load("accuracy")
metric.compute(predictions=preds, references=predictions.label_ids)
```
